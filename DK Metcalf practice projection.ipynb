```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Scrape DK Metcalf's receiving stats from Pro Football Reference
df = pd.read_csv('sportsref_download.csv')
```

```python
df.head()
```

Clean the data

```python
# Clean the data
df = df.dropna()
df = df.rename(columns={"G":"Games", "Tgt":"Targets", "Rec":"Receptions", "Yds":"Receiving_Yards", "Y/Tgt":"Yards_per_Target", "Y/G":"Yards_per_Game"})
```

```python
df.head()
```

```python
# Choose features
features = ["Age", "Games", "Receptions", "Targets", "Yards_per_Target", "Yards_per_Game"]
```

```python
# Train a linear regression model
model = LinearRegression()
model.fit(df[features], df["Receiving_Yards"])
```

```python
# Make a prediction for the upcoming season
age = 26
games = 17
receptions = 85
targets = 120
yards_per_target = 7.9
yards_per_game = 71.5

receiving_yards_prediction = model.predict([[age, games, receptions, targets, yards_per_target, yards_per_game]])[0]
```

```python
# Print the prediction
print(f"DK Metcalf's projected receiving yards for the upcoming season is {receiving_yards_prediction:,}")
```
